```julia
function get_sequences_from_file(file_name)
    num_sequences = 0
    
    # get the number of sequences from file
    open(file_name) do f
        sequence = ""
        for line in eachline(f)
            if startswith(line, 's') 
                line_to_array = split(line," ")
                index_of_sequence = parse(Int,string(line_to_array[1][2:end]))
                num_sequences = maximum([num_sequences,index_of_sequence])
            end
        end
    end
    
    sequences = Array{String,1}(undef,num_sequences)
    for i=1:length(sequences)
        sequences[i] = ""
    end
    
    # combine the sequences form file
    open(file_name) do f
        sequence = ""
        for line in eachline(f)
            if startswith(line, 's') 
                line_to_array = split(line," ")
                index_of_sequence = parse(Int,string(line_to_array[1][2:end]))
                sequence = line_to_array[end]
                sequences[index_of_sequence] = sequences[index_of_sequence]*sequence
            end
        end
    end
    return sequences
end

function calc_sum_pair(sequences) 
    t = length(sequences)
    k = length(sequences[1])
    score = 0
    for i=1:t
        A = sequences[i]
        for j=i+1:t
            B = sequences[j]
            for idx = 1:k
                if A[idx] == B[idx] && A[idx] != '-'
                    score += 1
                end
            end
        end
    end
    return score
end


function calc_sum_pair_include_gap_penalty(sequences) 
    t = length(sequences)
    k = length(sequences[1])
    score = 0
    for i=1:t
        A = sequences[i]
        for j=i+1:t
            B = sequences[j]
            for idx = 1:k
                # Add 1 for match
                if A[idx] == B[idx] && A[idx] != '-'
                    score += 1
                # subtract 1 for mismatch
                elseif A[idx] != B[idx] && A[idx] != '-' && B[idx] != '-'
                    score -= 1
                # subtract 2 for gap
                elseif A[idx] == '-' ||  B[idx] == '-' 
                    score -= 1
                end
            end
        end
    end
    return score
end

function get_score(filename)
    output_sequences = get_sequences_from_file(filename)
#     println(output_sequences)
#     println(length(output_sequences))
    t_coffee_score = calc_sum_pair_include_gap_penalty(output_sequences)
    return t_coffee_score
end

function driver(num_of_files_to_read)
    filename_starts_with = "output"
    filename_ends_with = ".fasta"
    scores = []
    for i=1:num_of_files_to_read
        filename = string(filename_starts_with,i,filename_ends_with)
        score = get_score(filename)
        push!(scores,score)
    end
    return scores
end

driver(11)
```
